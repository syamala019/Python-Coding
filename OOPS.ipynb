```python
class ECE:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    print(self.a+self.b)
  def subtraction(self):
    print(self.a-self.b)

  def common(self):
    self.addition()
    self.subtraction()

syamu = ECE(10,20)
syamu.common()
```

```python
class ECE:
  def __init__(self,a,b):
    self.a = a
    self.b = b

  def addition(self):
    self.f = 50
    print(self.a+self.b)

  def subtraction(self):
    print(self.a-self.b)

  def common(self):
    self.addition()
    self.subtraction()
    print(f'Accessing values : {self.f}')

syamu = ECE(10,20)
syamu.common()
```

**Inheritance**

Inheritance means => making the connection between the classes

They are 5 types of inheritance

* single inheritance
* multiple inheritance
* multilevel inheritance
* Hierarchical inheritance
* Hybrid inheritance

**single inheritance**

It have only one parent class and one child class

```python
class PARENT:
  def bussiness(self):
    print('Father is having bussiness')

class CHILD(PARENT):
  def job(self):
    print('I am having my own job')

syamu = PARENT()
syamu.bussiness()
```

```python
class PARENT:
  def bussiness(self):
    print('Father is having bussiness')

class CHILD(PARENT):
  def job(self):
    print('I am having my own job')

syamu = CHILD()
syamu.job()
syamu.bussiness()
```

**multilevel inheritance**

It have one grandparent class and one parent class and one child class

```python
 class GRANDFATHER:
  def building(self):
    print('I have my own building')

class FATHER(GRANDFATHER):
  def bussiness(self):
    print('Father is having bussiness')

class CHILD(FATHER):
  def job(self):
    print('I am having my own job')

syamu = CHILD()
syamu.job()
syamu.bussiness()
syamu.building()
```

**multiple inheritance**

It have two parent classes and one child class only

```python
class MOTHER:
  def building(self):
    print('I have my own building')

class FATHER:
  def bussiness(self):
    print('Father is having bussiness')

class CHILD(FATHER,MOTHER):
  def job(self):
    print('I am having my own job')

syamu = CHILD()
syamu.job()
syamu.bussiness()
syamu.building()
```

**Hierarchical inheritance**

It have one parent class and more than two child classes

```python
class FATHER:
  def college(self):
    print('As a father i am having college with me')

class B1(FATHER):
  def building(self):
    print('I have my own building')

class B2(FATHER):
  def bussiness(self):
    print('Father is having bussiness')

class B3(FATHER):
  def job(self):
    print('I am having my own job')

syamu = FATHER()
syamu.college()
```

```python
class FATHER:
  def college(self):
    print('As a father i am having college with me')

class B1(FATHER):
  def building(self):
    print('I have my own building')

class B2(FATHER):
  def bussiness(self):
    print('Father is having bussiness')

class B3(FATHER):
  def job(self):
    print('I am having my own job')

syamu = B2()
syamu.bussiness()
syamu.college()
```

**Hybrid inheritance**

It have one grandparent class and two parent classes and one child class

```python
class GRANDFATHER:
  def college(self):
    print('As a father i am having college with me')

class FATHER(GRANDFATHER):
  def building(self):
    print('I have my own building')

class MOTHER(GRANDFATHER):
  def bussiness(self):
    print('Father is having bussiness')

class CHILD(FATHER,MOTHER):
  def job(self):
    print('I am having my own job')

syamu = CHILD()
syamu.job()
syamu.bussiness()
syamu.building()
syamu.college()
```

**Polymorphism**

Polymorphism word -> derived from Greek Language

Poly -> Many

Morphism -> Forms

Basically polymorphism is used for 2 things

**Method[Function] Overriding -> we need one parent class and one child class**

**Method[Function] Overloading**

```python
# Method Overriding

def fun():
  print('good morning')

def fun():
  print('good evening')

fun()
```

```python
# if we want to implement -> method overriding concept -> we need one parent class and one child class for sure

class MANAGER:
  def __init__(self):
    self.a = 100

  def database_connection(self):
    print(f'The Address of a is : {id(self.a)}')

class MODULE_LEAD(MANAGER):
  def database_connection(self):
    self.b = self.a
    self.b = self.b + 100
    print(f'Issue was update here : {id(self.b)} : {self.b}')

syamu = MODULE_LEAD()
syamu.database_connection()
```

```python
# Method Overloading

def fun(a=None,b=None):
  if a==None and b==None:
    print('I am from India')
  elif a==None and b!=None:
    print('I am from USA')
  elif a!=None and b==None:
    print('I am from canada')
  elif a!=None and b!=None:
    print('I am from paris')
  else:
    print('I am from some other location')

fun(10,11)
```

**Abstraction**

If you want to use them after some time your class should be abstractclass and if you want to use the method like that ,that method should be abstractmethod

```python
from abc import ABC,abstractmethod
```

```python
class WHATSAPP(ABC):
  @abstractmethod
  def status(self):
    pass
  def chat(self):
    print(f'used for chatting')
  def calling(self):
    print(f'used for calling')

class CHILD(WHATSAPP):
  def status(self):
    print(f'I will write some logics for whatsapp here')

b = CHILD()
b.status()
```
