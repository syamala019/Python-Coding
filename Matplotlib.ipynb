**Matplotlib**

The purpose of using Matplotlib is to convert the data in the form of EDA [Exploratory Data Analysis]

**some visualizations we are going to use in Matplotlib-Pyplot**

* Scatter plot
* Line plot
* Legend
* Histogram
* Subplots
* Bar charts
* pie charts
* reading images

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Scatter_plot**

```python
a = [4,7,2,9,1,6]
b = [1,9,4,2,5,8]

plt.figure(figsize=(5,3)) # (w,h)

plt.scatter(x = a , y = b)

plt.show()
```

```python
a = [4,7,2,9,1,6]
b = [1,9,4,2,5,8]

plt.figure(figsize=(5,3))

plt.title('Scatter_plot_with_own_Data')
plt.xlabel('Data_Belongs to A')
plt.ylabel('Data_Belongs to B')

plt.scatter(x = a , y = b,color='g',marker='d')

plt.show()
```

```python
a = [4,7,2,9,1,6]
b = [1,9,4,2,5,8]

plt.figure(figsize=(5,3))

plt.xticks([])
plt.yticks([])

plt.scatter(x=a, y=b, color='purple', marker='d')
plt.show()
```

**Line plot**

```python
a = [4,7,2,9,1,6]
b = [1,9,4,2,5,8]

plt.figure(figsize=(5,3))

plt.title('Scatter_plot_with_own_Data')
plt.xlabel('Data_Belongs to A')
plt.ylabel('Data_Belongs to B')

plt.plot(a,b,marker='*')

plt.show()
```

```python
# drawback of line plot

india = [4,7,2,9,1,6]
usa = [1,9,4,2,5,8]
uk = [2,4,8,6,1,9]

plt.figure(figsize=(5,3))

plt.title('Scatter_plot_with_own_Data')
plt.xlabel('Data_Belongs to A')
plt.ylabel('Data_Belongs to B')

plt.plot(india,usa,marker='*')
plt.plot(india,uk,marker='*')

plt.show()
```

**Legend**

```python
india = [4,7,2,9,1,6]
usa = [1,9,4,2,5,8]
uk = [2,4,8,6,1,9]

plt.figure(figsize=(5,3))

plt.title('Scatter_plot_with_own_Data')
plt.xlabel('Data_Belongs to A')
plt.ylabel('Data_Belongs to B')

plt.plot(india,usa,marker='*',label='India_USA')
plt.plot(india,uk,marker='*',label='India_UK')

plt.legend(loc=1,fontsize='xx-large')

plt.show()
```

**Histogram -> it is used to find the data points in the given intervals(bins)**

```python
age = [1,4,9,3,6,11,14,17,22,23,29,28,32,33]

plt.figure(figsize=(5,3))

plt.hist(age,bins=3)

plt.show()
```

**Subplots**

subplot(no.of rows,no. of columns,position)

```python
india = [4,7,2,9,1,6]
usa = [1,9,4,2,5,8]
uk = [2,4,8,6,1,9]

plt.figure(figsize=(5,3))

plt.subplot(1,2,1)
plt.title('India-USA')
plt.plot(india,usa,color='r',marker='*')

plt.subplot(1,2,2)
plt.title('India-UK')
plt.plot(india,uk,marker='*')

plt.show()
```

```python
india = [4,7,2,9,1,6]
usa = [1,9,4,2,5,8]
uk = [2,4,8,6,1,9]

plt.figure(figsize=(3,3))

plt.subplot(2,1,1)
plt.title('India-USA')
plt.plot(india,usa,color='r',marker='*')

plt.subplot(2,1,2)
plt.title('India-UK')
plt.plot(india,uk,color='g',marker='*')

plt.show()
```

**Bar_chart**

```python
company  = ['Google','MSFT','Meta','IBM','Infosys']
revenue = [1200,1450,900,400,60]
plt.figure(figsize=(6,3))
plt.bar(company,revenue,color=['r','b','g','y','black'])

plt.show()
```

```python
company  = ['Google','MSFT','Meta','IBM','Infosys']
revenue = [1200,1450,900,400,60]
plt.figure(figsize=(6,3))
plt.barh(company,revenue,color=['r','b','g','y','black'])

plt.show()
```

**Piechart**

```python
plt.figure(figsize=(6,3))

company  = ['Google','MSFT','Meta','IBM','Infosys']
revenue = [1200,1450,900,400,60]

plt.pie(revenue,labels=company)

plt.show()
```

```python
plt.figure(figsize=(6,3))

company  = ['Google','MSFT','Meta','IBM','Infosys']
revenue = [1200,1450,900,400,60]

plt.pie(revenue,labels=company,shadow=True)

plt.show()
```

```python
plt.figure(figsize=(6,3))

company  = ['Google','MSFT','Meta','IBM','Infosys']
revenue = [1200,1450,900,400,60]

plt.pie(revenue,labels=company,autopct='%.3f')

plt.show()
```

```python
plt.figure(figsize=(6,3))

company  = ['Google','MSFT','Meta','IBM','Infosys']
revenue = [1200,1450,900,400,60]

plt.pie(revenue,labels=company,autopct='%.3f',explode=[0,0,0.2,0,0.4])

plt.show()
```
