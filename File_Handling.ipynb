**File_Handling** -> doing work on files.

File :- file is the place where we have the data.

** The purpopse of file handling**

work with the files ex:-> create a file and insert data or read data from the file

* w -> write
* r -> read
* a -> add

**w - Mode**

* 1. If the file is not there in the specific location mode w will create the file and will write the data into the file
* 2. If the file is there  and there is no data in the file means w will write the data into the file
* 3. If the file is there and there is data in the file means -> it will override the data

```python
a = open('syamu.txt','w') # w is created file and open the file

a.write('I am from India')
a.close()
```

```python
a = open('prasad.txt','w') # w is created file and open the file

a.write('I am from USA')
a.close()
```

```python
a = open('vasu.txt','w') # w is created file and open the file


a.write('Hyderabad')
a.close()
```

**a - Mode**

* 1. If the file is not there in the specific location mode a will create the file and will add the data into the file
* 2. If the file is there  and there is no data in the file means a will add the data into the file
* 3. If the file is there and there is data in the file means -> it will override the data

```python
a = open('syamu.txt','w') # w is created file and open the file

a.write('I am from India')
a.close()
```

```python
a = open('prasad.txt','w') # w is created file and open the file

a.write('I am from USA')
a.close()
```

```python
a = open('vasu.txt','w') # w is created file and open the file


a.write('Hyderabad')
a.close()
```

**r - Mode**

If the file is not there means it is not possible to read the data because file is not there
 * read -> will read the data exactly how it is in the file
 * readline -> read only few lines from the file
 * readlines -> read all the data from the file but in a list manner

```python
a = open('vasu.txt','r')
print(a.read())
a.close()
```

```python
a = open('vasu.txt','r')
print(a.read())
a.close()
```

```python
a = open('vasu.txt','r')
print(a.readline())
a.close()
```

```python
a = open('vasu.txt','r')
print(a.readlines())
a.close()
```

```python
# why to close the file

a = open('vasu.txt','w')
a.write('I am from India')
```

```python
# file handling in real time code

with open('vasu.txt','w') as a:
  a.write('I am from India')
```

```python
with open('vasu.txt','r') as a:
  print(a.read())
```
