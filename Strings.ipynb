```python
# capitalilze -> it will convert the starting letter into capital and remaining all small if aleady capital means it remains capital only

a = 'syamala ratnam'
a = a.capitalize()
print(a)
```

```python
a = 'data science is going to rule the world in future'
a = a.capitalize()
print(a)
```

```python
# title -> each word first letter will be capitalized and remaianing all small only

a = 'data science is going to rule the world in future'
print(a.title())
```

```python
# lower -> will convert capital letters to small letters

a = 'SYAMALA'
print(a.lower())
```

```python
# upper -> will convert small letters to capital letters

a = 'syamala'
print(a.upper())
```

```python
# islower -> checks all are small or not

a = 'syamala'
print(a.islower())
```

```python
# isupper -> checks all are capital or not

a = 'syamala'
print(a.isupper())
```

```python
# isalpha -> it will check all are alphabets or not

a = 'syamala'
print(a.isalpha())
```

```python
# isnumeric -> it will check all are numbers or not

a = "123"
print(a.isnumeric())
```

```python
# isalnum -> accepts numbers | alphabets | both

a = 'syamala19'
print(a.isalnum())
```

```python
# startswith

a = 'data science is going to rule the world in future'
print(a.startswith('data'))
```

```python
# endswith

a = 'data science is going to rule the world in future'
print(a.endswith('e'))
```

```python
# replace

a = 'data science'
a = a.replace('a','j',3)
print(a)
```

```python
# count

a = 'data science'
print(a.count('a'))
```

```python
# index

a = 'data science'
print(a.index('a'))
```
