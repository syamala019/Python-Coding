```python
# to optimize the error -> we need Exceptional handling concepts

# In Exceptional handling concepts => we have 2 things [blocks]

# try block and except block
```

```python
# try -> write the code in the try block  => if there is no issues in the code it execute
# except -> if there is any issue in the try block except block will work
```

```python
try:
  a = 10
  b = 20
  print(a / b)
except:
  pass
```

```python
try:
  a = 10
  b = 0
  print(a / b)
except Exception as e:
  print(e)
```

```python
try:
  city = 'delhi'
  print(CITY)

except Exception as e:
  print(e)
```

```python
try:
   a = 10
   b = 0
   print(a / b)

except ValueError as e:
  print(e)
except NameError as e:
  print(e)
```

```python
# finally -> it will generate with or without exeption

try:
   a = 10
   b = 20
   print(a / b)

except Exception as e:
  print(e)
finally:
  print('I will execute without error also')
```

```python
 try:
   a = 10
   b = 0
   print(a / b)

 except Exception as e:
  print(e)
finally:
  print('I will execute with error also')
```

```python
try:
  a = 10
  b = 0
  print(a / b)

except Exception as e:
  print(f'Issue generated from try block was : {e}')
```

```python
import sys
```

```python
try:
  a = 10
  b = 0
  print(a / b)

except Exception as e:
  print(f'Issue generated from try block was : {e}')
  error_type,error_msg,line_no = sys.exc_info()
  print(f'Error Type from sys library : {error_type}')
  print(f'Error Msg from sys library : {error_msg}')
  print(f'Line No: {line_no.tb_lineno}')
```

```python
try:
  a = 10
  b = 0
  print(a / b)

except Exception as e:
  error_type,error_msg,line_no = sys.exc_info()
  print(f'Error -> {error_type} Reason : {error_msg} in Line No : {line_no.tb_lineno}')
```
