**Rules to create a Varibale**

- `Rule no : 1 -> Varibale Should not start with a Number if you want to use Please use underscore _ `

- `Rule no : 2 -> Varibale should be single word or character -> Multiple words or multiple charaters are not allowed as a varibale if you want please use underscore _ `

- `Rule no : 3 -> Python Varibales are case sensitive ....-> [case sensitive -> capital letters and small letters ]`

- `Rule no : 4 -> Dont use these symbols as a varibale -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~ `

```python
# rule no :1

10a = 200 # save in the Memory

print(10a) # calling from the Memory
```

* step_3:it works with shortcuts -> int %d, float %f, str %s
