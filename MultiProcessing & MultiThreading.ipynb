```python
import time
import multiprocessing
import os

starting_time = time.time()

def india():
  print(f'India -> Processor Id:  {os.getpid()}')
  for i in range(1,11):
    time.sleep(0.5)
    print(f'India -> {i}')

def usa():
  print(f'USA -> Processor Id: {os.getpid()}')
  for j in range(51,61):
    time.sleep(0.5)
    print(f'usa -> {j}')

if __name__ == "__main__":
  print(f'Main processor id: {os.getpid()}')

  cpu1 = multiprocessing.Process(target=india,args=())
  cpu2 = multiprocessing.Process(target=usa,args=())

  cpu1.start()
  cpu2.start()

  cpu1.join()
  cpu2.join()

  print(f'Total time : {time.time() - starting_time}')
  print(f'Main processor exists ')
```

```python
#Multiprocessing

import time
import multiprocessing
import os

starting_time = time.time()

def india():
  print(f'India -> Processor Id:  {os.getpid()}')
  for i in range(1,11):
    time.sleep(0.5)
    print(f'India -> {i}')

def usa():
  print(f'USA -> Processor Id: {os.getpid()}')
  for j in range(51,61):
    time.sleep(0.5)
    print(f'usa -> {j}')

def common():
  india()
  usa()

if __name__ == "__main__":
  print(f'Main processor id: {os.getpid()}')

  cpu1 = multiprocessing.Process(target=common,args=())

  cpu1.start()

  cpu1.join()

  print(f'Total time : {time.time() - starting_time}')
  print(f'Main processor exists ')
```
