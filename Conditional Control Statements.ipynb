**Find the given number is a perfect number or not**

```python
n = 8
c = 0

for i in range(1,n):
  if n % i == 0:
    c = c + i

if n == c:
  print(f'{n} is perfect number')
else:
  print(f'{n} is not prim number')
```

**some logics**

```python
n = 8
c = 0
for i in range(1,n+1):
  if n % i == 0:
    c = c + 1

if c == 2:
  print(f'{n} is prime number')
else:
  print(f'{n} is not prime number')
```

```python
# 1 to 100 -> prime numbers

for i in range(1,101):
  c = 0
  for j in range(1,i+1):
    if i % j == 0:
      c = c + 1
  if c == 2:
    print(f'{i} is prime number')
  else:
    print(f'{i} is not prime number')
```

```python
# palindrome

a = 121
b = int(str(a)[::-1])
if a == b:
  print(f'{a} is palindrome')
else:
  print(f'{a} is not plindrome')
```

```python
n = 121
temp = n
c = 0

while n > 0:
  res = n % 10
  c = (c*10) + res
  n = n // 10

if temp == c:
  print(f'{temp} is palindrome')
else:
  print(f'{temp} is not plindrome')
```
