```python
for i in range(11,21):
  if i % 2 == 0:
    print(f'Even : {i}')
```

```python
# using functions

def syamu(a,b):
  for i in range(a,b):
    if i % 2 == 0:
      print(f'Even : {i}')

syamu(11,21)
```

```python
syamu(31,41)
```

```python
g = 19   # global variable
def fun():
  print(f'I am inside the function : {g}')

fun()
print(f'I am outside the function : {g}')
```

```python
def syamu():
  s = 19  # local variable
  print(f'I am inside the function : {s}')

syamu()
print(f'I am outside the function : {s}')
```

```python
# There is a chance to make local variable into global variable

def syamu():
  global s   # we know s is a local variable we are converting local into global
  s = 19     # local variable
  print(f'I am inside the function : {s}')

syamu()
print(f'I am outside the function : {s}')
```

```python
# return in python functions

def vasu(c,d):
  g = []
  for i in range(c,d):
    if i % 2 == 0:
      g.append(i)

  return g

sol = vasu(11,21)
print(sol+[10])
```

```python
n = 8
c = 0

for i in range(1,n):
  if n % i == 0:
    c = c + i

if n == c:
  print(f'{n} is perfect number')
else:
  print(f'{n} is not perfect number')
```

```python
def appu(n):
  c = 0
  for i in range(1,n):
    if n % i == 0:
      c = c + i

  if n == c:
    print(f'{n} is perfect number')
  else:
    print(f'{n} is not perfect number')

appu(8)
```

```python
appu(6)
```

**Lambda Function**

syntax : lambda arguments : expressions

```python
def fun(a,b):
  print(a+b)
fun(10,5)
```

```python
c = lambda x,y : x+y
print(c(10,5))
```

```python
print(c(1,5))
```

```python
n = 2
for i in range(1,11):
  print(f'{n} * {i} = {n*i}')
```

```python
# with function

def fun(n):
  for i in range(1,11):
    print(f'{n} * {i} = {n*i}')

fun(2)
```

```python
# with lambda function
c = lambda x,y : x*y

n = 2
for i in range(1,11):
  print(f'{n} * {i} = {c(n,i)}')
```
