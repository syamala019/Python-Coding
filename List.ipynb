**Python list is mutable or immutable -> prove it**

```python
# trying to add the memory space to the existing memory

a = [4,10,5,14,19] # save in the memory
a[5] =100 # trying to create a new index box and want to fill 100 in it
print(a)
```

List is Mutable only -> but we should not use index to increase or decrease

using the List built_in_functions it is possible

**Built_in_functions in list**

**addition**

* append
* insert
* extend

**deleting**

* pop
* remove

**to do some other normal operations**

* count
* index
* copy
* reverse
* sort
* clear

```python
# append -> it will take only one value from the user and will add to the existing memory only at the end
a = [4,10,5,14,19]
a.append(100)
print(a)
```

```python
a = [4,10,5,14,19]
a.append(100,200)
print(a)
```

```python
# extend -> will take more inputs from user and will add only to the end of the list
a = [4,10,5,14,19]
a.extend([100,200])
print(a)
```

```python
# insert -> we can add values where we want
a = [4,10,5,14,19]
a.insert(2,100)
print(a)
```

```python
# deleting
# pop -> pop defaulty will remove last value but if the user passing one value it will consider as a index
a = [4,10,5,14,19]
a.pop()
print(a)
```

```python
a = [4,10,5,14,19]
a.pop(2)
print(a)
```

```python
# remove -> will take the value from the user and will remove it
a = [4,10,5,14,19]
a.remove(14)
print(a)
```

so from the operations we can say that Python List is completely Mutable

```python
# count
a = [10,20,30,40,10,30,60]
print(a.count(10))
```

```python
# index
a = [10,20,30,40,10,30,60]
print(a.index(10))
```

```python
print(a.index(10,2))
```

```python
# reverse -> [::-1]
a = [10,20,30,40,10,30,60]
print(a[::-1])
```

```python
a = [4,10,5,14,19]
a.reverse()
print(a)
```

```python
# sort -> ascending order and descending order -> default -> ascending order
a = [4,10,5,14,19]
a.sort() # ascending order
print(a)
```

```python
a = [4,10,5,14,19]
a.sort(reverse=True) # descending order
print(a)
```

```python
# copy
# There are 2 types of copy
# 1.Deep copy and 2. Shallow copy
```

```python
# deep copy
a = [10,20,30,4]
b = a

print(id(a))
print(id(b))

a.append(100)
print(a)
print(b)
```

```python
# shallow copy
a = [10,20,30,4]
b = a.copy()

print(id(a))
print(id(b))

a.append(100)
print(a)
print(b)
```

```python
# clear
a = [10,20,30,4]
a.clear()
print(a)
```

```python
a = [10,20,30,4]
del a
print(a)
```
