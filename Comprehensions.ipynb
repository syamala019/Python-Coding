In the comprehension syntaxs

- `1. forloop -> [<printing_area> forloop]`

- `2. forloop and if -> [<printing_area> forloop if<condition>]`

- `3. forloop if and else -> [<print> if else <print> forloop]`

**List Comprehension**

```python
# In normal way
for i in range(11,17):
  print('good morning',end='  ')
```

```python
print(['good morning' for i in range(11,17)])
```

```python
# In normal way
for i in range(11,17):
  if i % 2 == 0:
    print(f'Even : {i}')
```

```python
print(['Even :'+str(i) for i in range(11,17) if i % 2 == 0])
```

```python
# In normal way
for i in range(11,17):
  if i % 2 == 0:
    print(f'Even : {i}')
  else:
    print(f'Odd : {i}')
```

```python
print([f'Even : {i}' if i%2 == 0 else f'Odd : {i}' for i in range(11,17)])
```

**Dictionary Comprehension**

```python
fruits = {'apple':100, 'banana':200, 'carrot':300}
for i in fruits:
  fruits[i] = fruits[i] + 10
print(fruits)
```

```python
fruits = {'apple':100, 'banana':200, 'carrot':300}
print({i:fruits[i]+10 for i in fruits})
```

```python
fruits = {'apple':100, 'banana':200, 'carrot':300}
for i in fruits:
  if fruits[i] > 100:
    fruits[i] = fruits[i]*3
print(fruits)
```

```python
fruits = {'apple':100, 'banana':200, 'carrot':300}
print({i:fruits[i]*3 for i in fruits if fruits[i]>100})
```

```python
fruits = {'apple':100, 'banana':200, 'carrot':300}
for i in fruits:
  if fruits[i] > 100:
    fruits[i] = fruits[i]*3
  else:
    fruits[i] = fruits[i]*2
print(fruits)
```

```python
fruits = {'apple':100, 'banana':200, 'carrot':300}
print({i:(fruits[i]*3 if fruits[i]>100 else fruits[i]*2) for i in fruits})
```
