**Sets**

* Python sets are going to represented with -> { }
* Python sets are unordered pair.
* In sets concept index will not there -> since index is not there -> slicing and skipping is also not possible in sets concept.
* Python sets won't allow duplicate values.
* Python sets are Mutable -> but it won't allow Mutable data types inside it.

```python
a = {10,20,30,40,50}
print(type(a))
```

```python
# Memory allocation
print(a)
```

```python
# indexing
print(a[1])
```

```python
# duplicate
a = {10,20,30,40,50,10,20}
print(a)
```

```python
a = {10,55.55,'data',(1,2,3)}
print(a)
```

```python
a = {10,55.55,'data',[1,2,3]}
print(a)
```

using set built_in_functions we can prove that sets are Mutable

**built_in_functions**

**addition**

* add
* update

**deleting**

* pop
* discard
* remove

**some other normal operations**

* clear
* union
* intersect
* copy

```python
# add -> add will take one value from the user and will add to the memory in random place
a = {10,20,30,100,200,150,80}
print(f'Before adding : {a}')
a.add(500)
print(f'After adding : {a}')
```

```python
# update -> will take multiple values and will add randomly
a = {10,20,30,100,200,150,80}
print(f'Before adding : {a}')
a.update([1000,2000])
print(f'After adding : {a}')
```

```python
# pop -> will remove one random value
a = {10,20,30,100,200,150,80}
print(f'Before removing : {a}')
a.pop()
print(f'After removing : {a}')
```

```python
# discard -> it will take a value and if the value is there means it will remove or else it will be silent
a = {10,20,30,100,200,150,80}
print(f'Before removing : {a}')
a.discard(100)
print(f'After removing : {a}')
```

```python
a = {10,20,30,100,200,150,80}
print(f'Before removing : {a}')
a.discard(1)
print(f'After removing : {a}')
```

```python
# remove -> it will take a value and if the value is there means it will remove or else it will through error
a = {10,20,30,100,200,150,80}
print(f'Before removing : {a}')
a.remove(20)
print(f'After removing : {a}')
```

```python
a = {10,20,30,100,200,150,80}
print(f'Before removing : {a}')
a.remove(2)
print(f'After removing : {a}')
```

according to the above operations we can say that set is a Mutable


```python
# copy
# shallow copy
a = {10,20,30,100,200,150,80}
b = a.copy()
print(id(a))
print(id(b))

a.add(1000)
print(a)
print(b)
```

```python
# deep copy
a = {10,20,30,100,200,150,80}
b = a
print(id(a))
print(id(b))

a.add(1000)
print(a)
print(b)
```

```python
# union
a ={1,2,3,4,5,6}
b = {5,6,7,8,9,10}
print(a.union(b))
```

```python
# intersect
print(a.intersection(b))
```
