**Dictionary**

* Anything we write in {} -> a set
* Anything we write in {} -> but in a key value pair is becomes dictionary
* here key -> means column name and value means -> data inside the column
* dict will prepare the data but it will not show the exsheet infront of you
* In the dict concept -> there will be no index -> no slicing -> no skipping

```python
a = {10,20,30,40}
print(type(a))
```

```python
a = {'Age':30,'Salary':12.55,'Location':'Hyderabad'}
print(type(a))
```

```python
print(a)
```

```python
a = {'Age':[10,20,30],'Salary':[120.55,150.55,190.86],'Location':['Hyd','Pune','Chennai']}
print(a)
```

**Dictionary is Mutable or Immutable -> Prove it**

```python
# adding
fruits = {'apple':100,'carrot':300}
print(f'Before : {fruits}')

fruits['apple'] = 500
print(f'After : {fruits}')
```

```python
fruits = {'apple':100,'carrot':300}
print(f'Before : {fruits}')

fruits['banana'] = 500
print(f'After : {fruits}')
```

```python
fruits= {}
print(f'Before : {fruits}')
fruits['apple'] = 900
fruits['cherry'] = 500
print(f'After : {fruits}')
```

```python
# deleting
a = {'apple':100,'carrot':300,'banana':500}
print(f'Before : {a}')
del a['carrot','banana']
print(f'After : {a}')
```

```python
a = {'Age':[10,20,30],'Salary':[120.55,150.55,190.86],'Location':['Hyd','Pune','Chennai']}
import pandas as pd
df = pd.DataFrame(a)
df = df.drop(['Salary','Location'],axis=1)
df
```

from the above operations we can say that -> dictionary is Mutable

**built_in_function_dict**

* get
* update
* pop
* popitem
* keys
* values
* items
* clear
* copy

```python
# get
fruits = {'apple':500,'carrot':200,'banana':600}
print(fruits['carrot'])
```

```python
print(fruits.get('carrot'))
```

```python
# update
fruits = {'apple':500,'carrot':200,'banana':600}
print(f'Before : {fruits}')
fruits['lemon'] = 1000
print(f'After : {fruits}')
```

```python
fruits = {'apple':500,'carrot':200,'banana':600}
print(f'Before : {fruits}')
fruits.update({'lemon':1000})
print(f'After : {fruits}')
```

```python
# pop -> will take key from the user and removes it
fruits = {'apple':500,'carrot':200,'banana':600}
print(f'Before : {fruits}')
del fruits['carrot']
print(f'After : {fruits}')
```

```python
fruits = {'apple':500,'carrot':200,'banana':600}
print(f'Before : {fruits}')
fruits.pop('carrot')
print(f'After : {fruits}')
```

```python
# popitem -> will reemove last key
fruits = {'apple':500,'carrot':200,'banana':600}
print(f'Before : {fruits}')
fruits.popitem()
print(f'After : {fruits}')
```

```python
#  key
fruits = {'apple':500,'carrot':200,'banana':600}
print(fruits.keys())
```

```python
# values
fruits = {'apple':500,'carrot':200,'banana':600}
print(fruits.values())
```

```python
# items
fruits = {'apple':500,'carrot':200,'banana':600}
print(fruits.items())
```

```python
# clear
fruits = {'apple':500,'carrot':200,'banana':600}
fruits.clear()
print(fruits)
```
