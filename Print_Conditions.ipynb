**Type Casting or Type Conversions**

Type Conversions means converting one data type into another data type

* int -> float and int -> str
* float -> int and float -> str
* str -> int and str -> float =>***

```python
# int-> float
a = 25
print(a)
print(type(a))

b = float(a)
print(b)
print(type(b))
```

```python
a = 55
str(a)
```

```python
# int -> str
a = 25 # saving in the memory
print(a) # calling from the memory
print(type(a)) # type is used to find that variable is holding which data type

b = str(a) # Actual conversion going here
print(b)
print(type(b))
```

```python
# float -> int
a = 25.55
print(a)
print(type(a))

b = int(a)
print(b)
print(type(b))
```

```python
# float -> str
a = 25.55
print(a)
print(type(a))

b = str(a)
print(b)
print(type(b))
```

```python
# str -> int
a = '25'
print(a)
print(type(a))

b = int(a)
print(b)
print(type(b))
```

```python
a = 'vizag'
print(a)
print(type(a))

b = int(a)
print(a)
print(type(b))
```

```python
# str -> float
a = '25'
print(a)
print(type(a))

b = float(a)
print(b)
print(type(b))
```

```python
a = 'chennai25.5'
print(a)
print(type(a))

b = float(a)
print(b)
print(type(b))
```

to make the relationship with one data type with other data type inside the  print -> to do this task we have 4 ways

* rule no.1 : using , comma  ->  we can add one data type with other data type and will also generate one tab space  ->  30%
* rule no.2 : using +    ->  we can add data types but make sure all are strings and also one tab space is not given
* rule no.3 : it works with shortcuts  ->  int %d , float %f , str %s
* rule no.4 : using f and {}  ->  real time using one

```python
print('My age is 12')
```

```python
# rule no.1 -> ,
a = 12
print('My age is',a)
```

```python
location = 'USA'
cost = 45000
ticket = 90000
print('I got college seat in',location,'which cost',cost,'and flight ticket was',ticket)
```

```python
# rule no.2 -> +
a = 12
print('My age is '+str(a))
```

```python
location = 'USA'
cost = 45000
ticket = 90000
print('I got college seat in '+location+ ' which cost '+str(cost)+ ' and flight ticket was '+str(ticket))
```

```python
# rule no.3 -> %d,%f,%s
a = 12
print('My age is %d' %(a))
```

```python
location = 'USA'
cost = 45000
ticket = 90000.55
print('I got college seat in %s which cost %d and flight ticket was %f'%(location,cost,ticket))
```

```python
# rule no.4 -> f and {}
a = 12
print(f'My age is {a}')
```

```python
location = 'USA'
cost = 45000
ticket = 90000
print(f'I got college seat in {location} which cost {cost} and flight ticket was {ticket}')
```
