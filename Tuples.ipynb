* Tuples are representing with -> ( )
* Tuple can accept any data type inside it.
* Memory allocation | indexing | slicing | skipping concepts are same like string and list.

```python
a = (10,55.55,'data') # any data type we can pass
print(type(a))
```

```python
# memory allocation
a = (10,11,20,14,9,6)
print(a)
```

```python
print(a[2])
```

```python
# slicing
print(a[2:4])
```

```python
print(a[2:])
```

```python
print(a[:4])
```

```python
print(a[:])
```

```python
print(a[::-1])
```

```python
# skipping
print(a[::2])
```

**Tuple is a Mutable or Immutable -> prove it**

* adding
* deleting

```python
# adding
a = (10,11,20,14,9,6)
a[6] = 100
print(a)
```

```python
# deleting
del a[6]
print(a)
```

From the above 2 operations tuple is an Immutable

**built_in_functions in tuple**

* count
* index

```python
# count
a = (10,11,20,14,9,6)
print(a.count(10))
```

```python
# index
a = (10,11,20,14,9,6)
print(a.index(20))
```

```python
# in strings there are 2 built_in_functions
# split and join
```

```python
# split
a = 'India is very good country'
print(a.split())
```

```python
a = 'India is very good.country'
a.split('.')
```

```python
# join
a = ['India','USA','UK']
' '.join(a)
```

```python
# question -> i.love.programming.language

# expected -> language.programming.love.i

a = 'i.love.programming.language'
b = a.split('.')

c = b[::-1]

d = '.'.join(c)
print(d)
```
