**Series**

Series is used to create 1 column representation which means it works only with 1D

```python
import numpy as np
import pandas as pd
```

```python
a = pd.Series([10,20,30,40])
a
```

```python
a = np.array([100,200,300,400]) # creating 1D data
b = pd.Series(a) # 1D data giving to series
b
```

```python
a = np.array([100,200,300,400])
b = pd.Series(a,index=['row1','row2','row3','row4'])
b
```

```python
b['row3']
```

**Data Frame**

It is used for multiple rows and multiple columns

```python
# using dictionary lets prepare the data

a = {'Name':['John','Peter','Micheal','Harry','Steve'],
      'Location':['USA','UK','Canada','Paris','India'],
      'Age':[22,25,29,30,33],
      'Salary':[1200,1500,2000,2100,4600],
      'Company':['google','msft','meta','IBM','Infosys']}
b = pd.DataFrame(a)
b
```

```python
b['Salary'][1]
```

```python
b['Location'][2]
```

```python
b['Location'][1]
```

loc and iloc are used to pull the selected area

* loc directly works with row and column as it is and no (n-1) logic
* iloc completely works with (n-1) logic and with index

  syntax: [r_s : r_e, c_s : c_e]

```python
# loc

b
```

```python
b.loc[2:3,'Location':'Salary']
```

```python
b.iloc[2:4,1:4]
```

```python
b.loc[1:2,'Name':'Location']
```

```python
b.loc[1:2,'Location':'Salary']
```

```python
# some more operations
b
```

```python
# rename the column

b = b.rename(columns={'Company':'Organisation'})
b
```

```python
# drop -> it is used to remove the column
# in pandas axis = 1 means columns and axis = 0 means rows

b = b.drop(['Age'],axis=1)
b
```

```python
b = b.drop([3],axis=0)
b
```
