**Python Operators**

operators -> which is use to do some mathematical operation like ->  +,-,/,//,*,pow,>,<,>=,<=,=,==,!=,and,or

* pow is a built in function in python.
*pow does not accept more than 3 arguments.

```python
# + -> addition

a=10
b=20
print(a+b)
```

```python
# - -> subtraction

a=10
b=20
print(a-b)
```

```python
# * -> multiplication

a=2
b=3
print(a*b)
```

```python
# / -> division

22/5
```

```python
# // -> floor division

22//5
```

```python
# % -> mod

22 % 5
```

```python
# ** -> power

2**5
```

```python
# pow
pow(2,5)
```

```python
pow(2,5,5)
```

```python
# >
20>10
```

```python
# <
20<10
```

```python
# >=
10>=9
```

```python
# <=
10<=9
```

```python
# =
a=10
print(a)
```

```python
# ==
10==9
```

```python
# !=
10!=5
```

```python
# and
10>=9 and 20<=28
```

```python
# or
10>=9 or 20<=19
```

```python
# \n -> it will go to the next line
print('syamala\nratnam')
```

```python
# \t -> it will give a big tab space
print('syamala\tratnam')
```

```python
print('\\n')
```
